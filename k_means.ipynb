```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


dataset = pd.read_csv('Mall_Customers.csv')
# convert the data in the second column to categorical data depending on the gender, male or female. The result will be a binary column
dataset.replace('Female', 0, inplace=True)
dataset.replace('Male', 1, inplace=True)
dataset.drop(["CustomerID"], axis = 1, inplace=True)


# visualize the data
plt.figure(figsize=(10,6))
plt.title("Ages Frequency")
sns.axes_style("dark")
sns.violinplot( y=dataset["Age"] )
plt.show()

```

```python

plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=dataset["Spending Score (1-100)"], color="red")
plt.subplot(1, 2, 2)
sns.boxplot(y=dataset["Annual Income (k$)"], color="green")
plt.show()
```

```python
#count the number of each gender
genders = dataset.Genre.value_counts()
sns.set_style("darkgrid")
plt.figure(figsize=(10,4))
sns.barplot(x=genders.index, y=genders.values)
plt.show()
```

```python
# create the conditions for the age groups
age18_25 = dataset.Age[(dataset.Age <= 25) & (dataset.Age >= 18)]
age26_35 = dataset.Age[(dataset.Age <= 35) & (dataset.Age >= 26)]
age36_45 = dataset.Age[(dataset.Age <= 45) & (dataset.Age >= 36)]
age46_55 = dataset.Age[(dataset.Age <= 55) & (dataset.Age >= 46)]
age55above = dataset.Age[dataset.Age >= 56]

x = ["18-25", "26-35", "36-45", "46-55", "55+"]
y = [len(age18_25.values), len(age26_35.values), len(age36_45.values), len(age46_55.values), len(age55above.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=x, y=y, palette="rocket")
plt.title("Number of Customers and Ages")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# create the conditions for the spending score groups
ss1_20 = dataset["Spending Score (1-100)"][(dataset["Spending Score (1-100)"] <= 20) & (dataset["Spending Score (1-100)"] >= 1)]
ss21_40 = dataset["Spending Score (1-100)"][(dataset["Spending Score (1-100)"] <= 40) & (dataset["Spending Score (1-100)"] >= 21)]
ss41_60 = dataset["Spending Score (1-100)"][(dataset["Spending Score (1-100)"] <= 60) & (dataset["Spending Score (1-100)"] >= 41)]
ss61_80 = dataset["Spending Score (1-100)"][(dataset["Spending Score (1-100)"] <= 80) & (dataset["Spending Score (1-100)"] >= 61)]
ss81_100 = dataset["Spending Score (1-100)"][(dataset["Spending Score (1-100)"] <= 100) & (dataset["Spending Score (1-100)"] >= 81)]

scoreX = ["1-20", "21-40", "41-60", "61-80", "81-100"]
scoreY = [len(ss1_20.values), len(ss21_40.values), len(ss41_60.values), len(ss61_80.values), len(ss81_100.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=scoreX, y=scoreY, palette="rocket")
plt.title("Number of Customers and Spending Scores")
plt.xlabel("Spending Score")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# create the conditions for the annual income groups
ai0_30 = dataset["Annual Income (k$)"][(dataset["Annual Income (k$)"] <= 30) & (dataset["Annual Income (k$)"] >= 0)]
ai31_60 = dataset["Annual Income (k$)"][(dataset["Annual Income (k$)"] <= 60) & (dataset["Annual Income (k$)"] >= 31)]
ai61_90 = dataset["Annual Income (k$)"][(dataset["Annual Income (k$)"] <= 90) & (dataset["Annual Income (k$)"] >= 61)]
ai91_120 = dataset["Annual Income (k$)"][(dataset["Annual Income (k$)"] <= 120) & (dataset["Annual Income (k$)"] >= 91)]
ai121_150 = dataset["Annual Income (k$)"][(dataset["Annual Income (k$)"] <= 150) & (dataset["Annual Income (k$)"] >= 121)]

annualIncomeX = ["0-30", "31-60", "61-90", "91-120", "121-150"]
annualIncomeY = [len(ai0_30.values), len(ai31_60.values), len(ai61_90.values), len(ai91_120.values), len(ai121_150.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=annualIncomeX, y=annualIncomeY, palette="rocket")
plt.title("Number of Customers and Annual Incomes")
plt.xlabel("Annual Income")
plt.ylabel("Number of Customers")
plt.show()
```
